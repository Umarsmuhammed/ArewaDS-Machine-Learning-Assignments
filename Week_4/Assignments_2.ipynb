# Exploring Visualizations
There are several different libraries that are available for data visualization. Create some visualizations using the Pumpkin data in this lesson with matplotlib and seaborn in a sample notebook. Which libraries are easier to work with?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pumpkins = pd.read_csv('US-pumpkins.csv')
pumpkins.head()
```

Relationship between Prices

```python
# scatter plot, matplotlib
plt.scatter('Low Price','High Price',data=pumpkins)
plt.title('Relationship between High and Low Prices')
plt.xlabel('Low Price')
plt.ylabel('High Price')
plt.show()
```

```python
# using matplotlib
plt.plot('Low Price','High Price',data=pumpkins)
plt.title('Relationship between High and Low Prices')
plt.xlabel('Low Price')
plt.ylabel('High Price')
plt.show()
```

```python
# using seaborn
sns.relplot(data=pumpkins,x='Low Price', y= 'High Price')
plt.title('Relationship between High and Low Prices')
plt.xlabel('Low Price')
plt.ylabel('High Price')
plt.show()
```
