# Visualizing relationship

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
honey = pd.read_csv('data/honey.csv')
honey.head()
```

```python
sns.relplot(x="year", y="stocks", kind="line", data=honey);
```

```python
honey['CA']=honey['state']
honey['CA']
```

```python
sns.relplot(x="stocks", y="year", hue="year", palette="Accent", data=honey, height=15, aspect=.5);
```

```python
sns.relplot(x="year", y="prodvalue", kind="line", data=honey);
```

Which State has seen highest volume in production, and in which year?

```python
honey[honey['totalprod']== honey['totalprod'].max()][["state","year"]]
```

What are the correlation between the numeric values

```python
sns.pairplot(honey[['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']])
```
