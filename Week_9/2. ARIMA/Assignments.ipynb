# ARIMA with Monthly Sunspot Data
we'll use the "Monthly Sunspot Numbers" dataset, which records the number of sunspots observed on the Sun each month from 1749 to 1983.
# 

```python
#importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#from statsmodels.tsa.arima.model import ARIMA
```

```python
# Load the Monthly Sunspot Numbers dataset
data = pd.read_csv('data/sunspots.csv')
data.head()
```

```python
data['Month'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)
```

```python
# Plot the original data
plt.figure(figsize=(10, 6))
plt.plot(data)
plt.xlabel('Year')
plt.ylabel('Sunspot Number')
plt.title('Monthly Mean total Sunspot Numbers Dataset')
plt.show()
```
